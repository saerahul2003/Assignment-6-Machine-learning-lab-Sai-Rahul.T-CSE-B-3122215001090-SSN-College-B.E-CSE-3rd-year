```python

from sklearn.cluster import KMeans
sse = []
start = 4
end = 11
for k in range(start,end):
    kmeans = KMeans(
        n_clusters=k,
        max_iter=10000,
    )
    kmeans.fit(df.iloc[::,:-2])
    sse.append(kmeans.inertia_)
plt.plot(range(start,end), sse)
plt.xticks(range(start,end))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()

diff = np.diff(sse)
print(diff)
```
